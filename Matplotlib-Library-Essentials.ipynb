### Introduction to Pyplot

```python
import matplotlib.pyplot as plt
import numpy as np

x_vals = np.linspace(0, 10, 1000)          # 1000 points on [0, 10]
y_vals = np.array([x**3 for x in x_vals])  # x**3

f = lambda x: x**2                         # x**2

plt.plot(x_vals, f(x_vals), 'b--', label='$x^2$')  # plot f(x) = x**2 as blue --
plt.plot(x_vals,   y_vals,  'g--', label='$x^3$')  # plot y_vals (x**3) as green --
```

```python
# we can easily add x and y axis labels

plt.xlabel('x (units)')
plt.ylabel('y (units)')

# and we can add a title and legend
plt.title('Plotting f(x) = $x^2$ vs. f(x) = $x^3$')
plt.legend()

plt.show()
```

###### Formatting the style of your plot

```python
x_vals = np.linspace(0, 10, 20)
squares = x_vals**2
cubes = x_vals**3

fig = plt.plot(x_vals, x_vals, 'ro')  # red ovals
fig = plt.plot(x_vals, squares, 'bs') # blue squares
fig = plt.plot(x_vals, cubes, 'g^')   # green triangles
fig = plt.show()
```

### Plotting with keyword strings

```python
data = {'x_vals': np.linspace(0, 100, 1000), 
        'y_vals': np.linspace(0, 100, 1000)**2, 
        'style':'b--'}

fig = plt.plot(data['x_vals'], data['y_vals'], data['style'], label='f(x) = $x^2$')
fig = plt.title('Plotting with dictionary keys')
fig = plt.xlabel('x')
fig = plt.ylabel('y')
fig = plt.legend()
fig = plt.show()
```

### Plotting with categorical variables

```python
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
car_accidents = [4603, 4455, 4508, 4767, 5865, 6826, 6007] # actual NHTSA data from 2000-2009

fig = plt.bar(days, car_accidents)
fig = plt.suptitle('*Actual NHTSA statistics from 2000-2009')
fig = plt.title('Average Yearly Accidents By Day')
fig = plt.xlabel('Day of Week')
fig = plt.ylabel('Accidents Per Year')
fig = plt.show()
```

### Controlling line properties

```python
# we can set linewidth, dash style, antialiased, and more

from math import sqrt

x_vals = np.linspace(0, 100, 1000)
y_vals = np.array([sqrt(x) for x in x_vals])

fig = plt.plot(x_vals, y_vals, linewidth=2.5, color='purple', linestyle='-.')
fig = plt.title('Plotting $\sqrt{x}$ on [0, 100]')
fig = plt.xlabel('x')
fig = plt.ylabel('y = $\sqrt{x}$')
fig = plt.axis([0, 100, 0, 25])
fig = plt.show()
```

### Working with multiple figures and axes

```python
x_vals = np.linspace(0, 4*np.pi, 30)

fig = plt.figure(1)

# Creating the first subplot

fig = plt.subplot(211)
fig = plt.title('Plotting sin(x)')
fig = plt.xlabel('x')
fig = plt.ylabel('y = sin(x)')
fig = plt.plot(x_vals, np.sin(x_vals), 'b^', x_vals, np.sin(x_vals), 'k')

# Creating the second subplot

fig = plt.subplot(212)
fig = plt.title('Plotting cos(x)')
fig = plt.xlabel('x')
fig = plt.ylabel('y = cos(x)')
fig = plt.plot(x_vals, np.cos(x_vals), 'r:')

# adjusting the spacing

fig = plt.subplots_adjust(top=1.0, bottom=0.08, left=0.10, right=0.95, hspace=0.5,
                    wspace=0.35)
fig = plt.show()
```

### Working with text

```python
mu, sigma = 100, 15
data = mu + sigma * np.random.randn(10000)

fig = plt.hist(data, bins=50)

plt.title('Data That Follows A Normal Distribution')
fig = plt.xlabel('Bins')
fig = plt.ylabel('Data')

fig = plt.text(50, 500, r'$\mu=100$, $\sigma=15$') # add text @ bin 50 and value 500
fig = plt.grid(True) # turn on grid

fig = plt.show()
```

###### Annotating text

```python
x = np.linspace(0, 2*np.pi, 100)

fig = plt.plot(x, np.sin(x))

fig = plt.title('Annotating Figure Example')

fig = plt.annotate('local max', xy=(1.5, 0.75), xytext=(1.5, 0),
             arrowprops=dict(facecolor='black', shrink=0.05),
             )
fig = plt.annotate('local min', xy=(4.5, -0.75), xytext=(5, 0),
             arrowprops=dict(facecolor='black', shrink=0.05),
             )
fig = plt.show()
```
